```python
class Posiłek:
    def __init__(self, nazwa, opis, waga, cena, porcje):
        self.nazwa = nazwa
        self.opis = opis
        self.waga = waga
        self.cena = cena
        self.porcje = porcje

    def __repr__(self):
        return f"{self.nazwa} - {self.opis}, Waga: {self.waga} dkg, Cena: {self.cena} zł, Ilość: {self.porcje} porcji"

    def dostępny(self, liczba):
        return self.porcje >= liczba

    def aktualizuj_porcje(self, liczba):
        if self.dostępny(liczba):
            self.porcje -= liczba
        else:
            raise ValueError(f"Brak wystarczającej ilości porcji dla {self.nazwa}")

    def dodaj_ilość(self, liczba):
        self.porcje += liczba
        print(f"Dodano {liczba} porcji dla posiłku {self.nazwa}")


class Zamówienie:
    def __init__(self):
        self.lista = []
        self.koszt = 0

    def dodaj_zamówienie(self, posiłek, ilość):
        if posiłek.dostępny(ilość):
            self.lista.append((posiłek.nazwa, posiłek.opis, posiłek.waga, posiłek.cena, ilość))
            self.koszt += posiłek.cena * ilość
            posiłek.aktualizuj_porcje(ilość)
        else:
            raise ValueError(f"Dostępnych {posiłek.porcje} porcji posiłku {posiłek.nazwa}")

    def __repr__(self):
        podsumowanie_zamówienia = "\n".join(
            [f"{nazwa} - {opis}, Waga: {waga} dkg, Cena: {cena} zł, Ilość: {ilość} porcji"
             for nazwa, opis, waga, cena, ilość in self.lista])
        return f"Podsumowanie zamówienia:\n{podsumowanie_zamówienia}\nCałkowity koszt: {self.koszt} zł"


class Jadłodajnia:
    def __init__(self):
        self.posiłki = []

    def dodaj_posiłek(self, posiłek):
        self.posiłki.append(posiłek)

    def wyświetl_posiłki(self):
        if not self.posiłki:
            print("Brak posiłków w jadłodajni")
        else:
            print("Jadłodajnia:")
            for posiłek in self.posiłki:
                print(posiłek)

    def zrób_zamówienie(self):
        zamówienie1 = Zamówienie()
        while True:
            self.wyświetl_posiłki()
            wybierz = input("Podaj nazwę posiłku (jeśli chcesz zakończyć, napisz 'wszystko'): ")
            if wybierz.lower() == "wszystko":
                break
            zamówienia_jadłodajni = [posiłek for posiłek in self.posiłki if posiłek.nazwa.lower() == wybierz.lower()]
            if zamówienia_jadłodajni:
                ilość = int(input(f"Podaj ilość porcji: "))
                try:
                    zamówienie1.dodaj_zamówienie(zamówienia_jadłodajni[0], ilość)
                    print(f"Dodano {ilość} porcji {wybierz} do zamówienia.")
                except ValueError as e:
                    print(e)
            else:
                print("Brak posiłków o tej nazwie w jadłodajni.")
        print(zamówienie1)


# Example usage:
if __name__ == "__main__":
    jadłodajnia = Jadłodajnia()

    posiłek1 = Posiłek("Zupa Pomidorowa", "Zupa z pomidorów z makaronem", 35, 12.50, 10)
    posiłek2 = Posiłek("Spaghetti", "Spaghetti z sosem pomidorowym i serem", 40, 18.00, 5)

    jadłodajnia.dodaj_posiłek(posiłek1)
    jadłodajnia.dodaj_posiłek(posiłek2)

    jadłodajnia.zrób_zamówienie()
```
